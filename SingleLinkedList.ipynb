```python
class SinglyLinkedList:
  """LIFO Stack implementation using a singly linked list for storage."""

 #-------------------------- nested _Node class --------------------------
  class _Node:         # "inner" class
    """Lightweight, nonpublic class for storing a singly linked node."""
    __slots__ = '_element', '_next'           # streamline memory usage

    def __init__(self, element, next):        # initialize node's fields
        self._element = element               # reference to user's element
        self._next = next                     # reference to next node

    def setNext(self, next):
        self._next = next
    
    def getNext(self):
        return self._next
  
    def setElement(self, element):
        self.element = _element
    
    def getElement(self):
        return self._element
        
  #------------------------------- stack methods -------------------------------
  def __init__(self, head = None, size = 0):
    """Create an empty list."""
    self._head = head                       # reference to the head node
    self._size = size                       # number of stack elements

  def getSize(self):
    """Return the number of elements in the list."""
    return self._size

  def isEmpty(self):
    """Return True if the list is empty."""
    return self._size == 0

  def add (self, element):
    """Add element element to the end of the list."""
    node = self._Node(element, None)
    if self.isEmpty():                       # special case if list is empty
        self._head = node
    else:
        cursor = self._head                  # list already has elements ...
        while cursor.getNext() is not None:  #   ...iterate to the end and add
            cursor = cursor.getNext( )
        cursor.setNext(node)
    print('Adding # ', self._size, ' element: ', node.getElement())
    self._size += 1
        
  def remove(self):
    """Remove the element at the head of the list. Raise Empty if the list is empty."""
    if self.isEmpty():
      raise Empty('List is empty')
    self._size -= 1
    out = self._head.getElement()      # save the head element and return it to the caller
    print("\nRemoving element: ",out, 'new size',self._size - 1,'\n')
    self._head = self._head.getNext()  # Reset the head to the second (next) list element
    return out

 # Added by GRS for in-class demo ----------------------------------------------------------------------
  def showList(self):
    """Display the contents of the list, from head to tail. Raise Empty exception if the List is empty. """
    cursor = self._head
    while cursor:
        print('List element: {0:15s}   object id: {1:15d}   next: {2:15d}'. \
              format(cursor.getElement(), id(cursor), id(cursor.getNext())))
        cursor = cursor.getNext()
    return
        
  def search(self, key):
    """ Search the list using 'key. """        
    cursor = self._head
    while cursor:
        if key == cursor.getElement():
            return True
        cursor = cursor.getNext()
    return False

    # Add Insert After
  def insertAfter(self,index,new_data): 
    print('\nEntering element ',new_data, ' after indexed element # ', index,'\n')
    new_node = self._Node(new_data, None)
    cursor = self._head
    counter = 0 # counter is 0 
    found = False
    pre = cursor
    self._size += 1
    
    # loop through linked list
    while cursor is not None: 
        if index == 0:     # if index is head 
            
            new_node.setNext(self._head)    # point new nodes next pointer to head
            self._head = new_node  # set head node to new node
            found = True    # found to true
            break
            
        if counter >= index - 1:  # if counter is greater than index minus 1 (cause list starts at 0)
            new_node.setNext(cursor.getNext())  # set new nodes next pointer to cursors next node
            cursor.setNext(new_node)   # set cursors next to new_node
         
            found = True  
            break
            
        pre = cursor
        cursor = cursor.getNext()
        counter += 1
       
    if found==False and index > 0:  # dont do it if we had special case
          
        pre.setNext(new_node)

# --------------------------------------------------------------------------------------
# insert after particular node instances
  def insertElt(self,new_data,insert,president_inList):
    
    cursor = self._head
    counter = 0 # counter is 0 
    found = False
    pre = cursor
    president = new_data
    new_node = self._Node(new_data,None)
   

    # if the the insert is After    
    if insert == "A":
        while cursor is not None: 
            if cursor._element == president_inList:
               
                found = True
                if found:
                    print("Adding # ",self._size,' element: ', new_data, ' after ', cursor._element)
                    new_node.setNext(cursor.getNext())  # point new node , curso
                    cursor.setNext(new_node) 
                    self._size += 1
                    break  
            
            cursor = cursor.getNext()
            
# --------------------------------------------------------------------------------------
    # if the insert is before
    elif insert == "B":
        cursor = self._head
        pre = cursor
        found = False
       
        
        
        
        if cursor == None:
            print('Nothing to insert before')  
            
        if cursor.getElement() == president_inList:
            new_node.setNext(self._head)
            self._head = new_node
            found = True
            self._size += 1
     
        else:
            # loop through linked list 
            while  cursor is not None:
                
                # if the crusor is equal to the given element in list                     
                if cursor._element == president_inList:
                    found = True
                    print("Adding # ",self._size,' element: ', new_data, 'before', cursor._element)
                    pre.setNext(new_node)  # point previous node to new node
                    new_node.setNext(cursor) # set new node to next 
                    self._size += 1
                    break
                pre = cursor
                cursor = cursor.getNext()
                
        # error handling 
        if not found:
            print("Sorry Error occured could not update")
                
    else:
        print("Sorry did not recive correct paramaters ")
    
# special case for head node
# special case for exeeding list just add to tail 
        
          
    
    

# Added by GRS for in-class demo -------------------------------------------------------------------
class Empty(Exception):
    """Empty exception class provided to flag that condition. """
    pass

```
